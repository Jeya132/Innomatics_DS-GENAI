```python
class Solution(object):
    def defangIPaddr(self, address):
        """
        :type address: str
        :rtype: str
        """
        result=''
        for i in address:
            if i != '.':
                result+=i
            else:
                result+="[.]"
        return result
```

```python
S=Solution()
S.defangIPaddr('1.1.1.1')
```

```python
class Solution(object):
    def findNumbers(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        result=0
        for i in nums:
            if len(str(i))%2==0:
                result+=1
        return result
```

```python
S=Solution()
S.findNumbers([12,345,2,6,7896])
```

```python
class Solution(object):
    def numIdenticalPairs(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        count = 0
        for i in range(len(nums)):
            for j in range(i + 1, len(nums)):
                if nums[i] == nums[j]:
                    count += 1
        return count
```

```python
S=Solution()
S.numIdenticalPairs([1,2,3,1,1,3])
```

```python
class Solution(object):
    def smallerNumbersThanCurrent(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        result=[]
        for i in nums:
            c=0
            for j in nums:
                if j<i and j!=i:
                    c+=1
            result.append(c)
        return result
```

```python
S=Solution()
S.smallerNumbersThanCurrent([8,1,2,2,3])
```

```python
class Solution(object):
    def subtractProductAndSum(self, n):
        """
        :type n: int
        :rtype: int
        """
        p=1
        s=0
        for i in str(n):
            p=p*int(i)
            s=s+int(i)
        return p-s
```

```python
S=Solution()
S.subtractProductAndSum(234)
```

```python
class Solution(object):
    def xorOperation(self, n, start):
        """
        :type n: int
        :type start: int
        :rtype: int
        """
        result = 0
        for i in range(n):
            result ^= start + 2 * i
        return result
```

```python
S=Solution()
S.xorOperation(5,0)
```
