```python
class Solution(object):
    def maxProduct(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        first = 0
        second = 0
        for num in nums:
            if num > first:
                second = first
                first = num
            elif num > second:
                second = num
        return (first - 1) * (second - 1)
```

```python
Solution().maxProduct([3,4,5,2])
```

```python
class Solution(object):
    def numTeams(self, rating):
        """
        :type rating: List[int]
        :rtype: int
        """
        n=len(rating)
        teams=0
        for j in range(1,n-1):
            ls=lg=rs=rg=0
            for i in range(j):
                if rating[i]<rating[j]:ls+=1
                else:lg+=1
            for k in range(j+1,n):
                if rating[k]>rating[j]:rg+=1
                else:rs+=1
            teams+=ls*rg+lg*rs
        return teams
```

```python
Solution().numTeams([2,5,3,4,1])
```

```python
class Solution(object):
    def busyStudent(self, startTime, endTime, queryTime):
        """
        :type startTime: List[int]
        :type endTime: List[int]
        :type queryTime: int
        :rtype: int
        """
        count=0
        for i in range(len(startTime)):
            if startTime[i]<=queryTime<=endTime[i]:
                count+=1
        return count
```

```python
Solution().busyStudent([1,2,3],[3,2,7],4)
```

```python
class Solution(object):
    def numberOfSteps(self, num):
        """
        :type num: int
        :rtype: int
        """
        steps=0
        while num>0:
            if num%2==0:
                num//=2
            else:
                num-=1
            steps+=1
        return steps
```

```python
Solution().numberOfSteps(14)
```

```python
class Solution(object):
    def countBits(self, n):
        """
        :type n: int
        :rtype: List[int]
        """
        ans=[0]*(n+1)
        for i in range(1,n+1):
            ans[i]=ans[i//2]+i%2
        return ans
```

```python
Solution().countBits(5)
```
